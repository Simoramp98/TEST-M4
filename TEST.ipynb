## 3) Sempre riguardo i casi di COVID totali, si chiede di sviluppare una funzione che prenda in input il dataset e due nomi di continenti, e che ne confronti i seguenti relativi descrittori statistici: valori minimo e massimo, media, e percentuale rispetto al numero di casi totali nel mondo (in questo caso calcolati anche sulle locazioni senza indicazione di continente):

```python
def funzione1(continent1, continent2):

    dati_continent = dati.loc[:, "continent":"new_cases":4]

    dati_continent1_filtro = dati_continent[dati_continent["continent"]==continent1]
    dati_continent1 = dati_continent1_filtro[dati_continent1_filtro["new_cases"]!=0]
    dati_continent1_cases = dati_continent1.loc[:, "new_cases"]
    dati_continent1_min_cases = dati_continent1_cases.min()
    dati_continent1_max_cases = dati_continent1_cases.max()
    dati_continent1_avg = dati_continent1_cases.mean()

    dati_continent2_filtro = dati_continent[dati_continent["continent"]==continent2]
    dati_continent2 = dati_continent2_filtro[dati_continent2_filtro["new_cases"]!=0]
    dati_continent2_cases = dati_continent2.loc[:, "new_cases"]
    dati_continent2_min_cases = dati_continent2_cases.min()
    dati_continent2_max_cases = dati_continent2_cases.max()
    dati_continent2_avg = dati_continent2_cases.mean()
    
    confronto_continenti = dati.groupby(by = "continent", as_index = False).agg(casi_totali = ("new_cases", np.sum))

    cf_continenti = dati.groupby(by = "continent", as_index = True).agg(casi_totali = ("new_cases", np.sum))

    numeri_confronto = confronto_continenti.loc[:, "casi_totali"].astype(int)
    totale_continent1 = cf_continenti.loc[continent1, :]
    totale_continent2 = cf_continenti.loc[continent2, :]
    totale_casi = numeri_confronto.sum()
    perc_continent1 = (totale_continent1 * 100)/totale_casi
    perc_continent2 = (totale_continent2 * 100)/totale_casi

    differenza_min = dati_continent1_min_cases - dati_continent2_min_cases
    differenza_max = dati_continent1_max_cases - dati_continent2_max_cases
    differenza_avg = dati_continent1_avg - dati_continent2_avg
    differenza_perc = perc_continent1 - perc_continent2

    confronto_dati = {"Min": [dati_continent1_min_cases, dati_continent2_min_cases, differenza_min], "Max": [dati_continent1_max_cases, dati_continent2_max_cases, differenza_max], "Media": [dati_continent1_avg, dati_continent2_avg, differenza_avg], "% su Tot": [perc_continent1, perc_continent2, differenza_perc]}
    confronto = pd.DataFrame(data=confronto_dati, index = [continent1, continent2, "Confronto"])
    print(confronto.round(0).astype(int))
```
